```python
drug_list = set(data['sm_name'].tolist())
drug_list
```

```python
# Go through list of drugs now + parse from openFDA source info about them
import requests

information = []
scrape_keys = [["pharm_class_cs", True], ["pharm_class_moa", False], ["pharm_class_epc", True], ["route", True]]
for drug in drug_list:
    print(F"Getting information about drug {drug}...")
    resp = requests.get(f"https://api.fda.gov/drug/label.json?api_key=EFOMPmU0YjFbeaWfXgEYAXClrPJrVq3YfrvPpuMD&search=openfda.generic_name:\"{drug}\"")
    info = resp.json()
    inf_drug = [drug]
    if "results" in info:
        # Attempt to find the info we need in here: info["results"][0]["openfda"]
        for key in scrape_keys:
            if key[0] in info["results"][0]["openfda"]:
                if key[1]:
                    inf_drug.append(info["results"][0]["openfda"][key[0]][0])
                else:
                    inf_drug.append(info["results"][0]["openfda"][key[0]])
            else:
                inf_drug.append(None)
            
        if "indications_and_usage" in info["results"][0]:
            inf_drug.append(info["results"][0]["indications_and_usage"][0])
        else:
            inf_drug.append(None)

        print(f"Adding drug information: {inf_drug}")
    else:
        inf_drug = [drug, None, None, None, None, None]
        
    information.append(inf_drug)
```

```python
information
```

```python
# Index JSON by drug
drug_inf = {}

for drug_i in information:
    drug_inf[drug_i[0]] = drug_i[1:]
```

```python
# Save all of this information to a JSON file
import json

with open('drug_information_indexed.json', 'w') as fp:
    json.dump(drug_inf, fp, sort_keys=True, indent=4)
```
